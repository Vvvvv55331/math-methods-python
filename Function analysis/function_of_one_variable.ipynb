```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rc('font', size=11)

x_min = 0.01
x_max = 10
x = np.linspace(x_min, x_max, 1000)

y1 = (x**3 + 3 * x**2) / 4 - 5
y2 = 2 / np.sin(0.1 * x)

plt.figure(figsize=(7, 4))

plt.plot(x, y1, label='y1(x)')
plt.plot(x, y2, label='y2(x)')

plt.xlabel('x')
plt.ylabel('y(x)')
plt.title(f'Графики функций y1(x) и y2(x) ' 
          f'на отрезке [{x_min}, {x_max}]',
          fontsize=12)

plt.legend()
plt.xlim(x_min, x_max)
plt.grid()

plt.show()
```

```python

def display_plot(points, x_scale):
    
    x = np.linspace(0.01, 10, points) 

    y1 = (x**3 + 3 * x**2) / 4 - 5
    y2 = 2 / np.sin(0.1 * x)

    y3 = np.sin(y1)
    y4 = np.sin(y2)

    plt.figure(figsize=(10, 2))

    plt.plot(x, y3, label='y3(x)')
    plt.plot(x, y4, label='y4(x)')

    plt.xlabel('x') 
    plt.ylabel('y3, y4') 

    if x_scale == 'log':
        scale_label = 'лог.' 
        plt.xscale('log') 
    else: 
        scale_label = 'лин.' 

    graph_title = (f'Графики функций y3(x) и y4(x) ' 
        f'{points} точек, {scale_label} масштаб')
    
    plt.title(graph_title) 
    plt.legend()
    plt.grid()

display_plot(600, 'lin')
display_plot(20000, 'lin')
display_plot(20000, 'log')
display_plot(500000, 'log')

plt.show()
```

```python

def y1(x):
    return (x**3 + 3 * x**2) / 4 - 5

def y2(x):
    return 2 / np.sin(0.1 * x)

x_all = np.linspace(0.01, 10, 10**6) 

y1_v = y1(x_all)
M = np.max(np.abs(y1_v))

y2_v = np.abs(y2(x_all))
idx_L = np.argmin(np.abs(y2_v - M))
L = x_all[idx_L]

x_range = np.linspace(L, 10, 1000)
y1_range = y1(x_range)
y2_range = y2(x_range)

def display_plot(x, y1, y2, M, x_scale):
    plt.figure(figsize=(8, 4))

    plt.plot(x, y1, label='y1(x)')
    plt.plot(x, y2, label='y2(x)')
    
    if x_scale == 'log':
        plt.xscale('log')
        scale_label = 'логарифмический масштаб'
    else:
        scale_label = 'линейный масштаб'

    plt.xlabel('x')
    plt.ylabel('y(x)')
    plt.title(f'Графики y1(x) и y2(x) на интервале [{L:.4f}'
              f', 10], {scale_label}',
              fontsize=11)
    
    plt.ylim(-M, M)
    plt.minorticks_on()
    plt.grid(which='major')
    plt.grid(which='minor', linestyle=':')
    plt.legend()

display_plot(x_range, y1_range, y2_range, M, x_scale='lin')
display_plot(x_range, y1_range, y2_range, M, x_scale='log')

plt.show()

print(f'M = {M:.2f}, L ≈ {L:.4f}')
```
